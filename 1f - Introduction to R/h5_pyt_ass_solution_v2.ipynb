**TASK 1**

Write a function named `play_mental_game` which takes two inputs - `game_attempts` : an integer representing the number of attempts at the game, and `crystals_collected` : an integer representing the number of crystals the player has collected so far.  The function should randomly choose two numbers between 1 and 10 using the `randint()` method from the `random` library.  Then, a for loop should be constructed that runs for the number of times specified in `game_attempts`.  In each run of the for loop, the computer should ask the user what the answer is to the two numbers randomly chosen being multiplied together (it should print a message to the user with the question too - e.g. `"What is 4 x 9?"`).  If the user answers correctly, a message should be displayed saying they got it correct, `crystals_collected` should be incremented by 1 and the the code should break out of the loop.  If the user answers incorrectly, a message should be printed telling them they are incorrect (they will get another go if there are attempts remaining).  At the end of the function, `crystals_collected` should be returned.

Note - for this task, you do not need to catch exceptions in the user input, although you may do so if you wish.

You can run the code in the code cell below the empty one to test whether your function is working.  Don't forget you must run your code cell after you make any changes for them to take effect.

```python
import random

def play_mental_game(game_attempts, crystals_collected):
    number_1 = random.randint(1,10)
    number_2 = random.randint(1,10)

    for i in range(game_attempts):
        print (f"What is {number_1} x {number_2}?")
        player_answer = int(input("ANSWER : "))

        if player_answer == number_1 * number_2:
            print ("Correct!")
            crystals_collected += 1
            break
        else:
            print ("Wrong!")

    return crystals_collected
```

```python
current_crystals = 5

new_number_of_crystals = play_mental_game(3, 5)

if new_number_of_crystals == current_crystals:
    print ("You've been locked in - sorry!")
else:
    print ("Congratulations - you won a crystal!")
```

**TASK 2**

Write a class named `Player` that has the following :
* a constructor that takes in a name, age and location, and which assigns name, age and location attributes to the passed in values.  It should also set an attribute named `locked_in` to the Boolean value `False`
* a method named `lock_in`, which sets the `locked_in` attribute of the Player object to `True`
* a method named `be_released`, which accepts a variable named `crystals_collected`, and which checks whether `crystals_collected` is greater than 0.  If it is, then `1` should be deducted from the number of crystals collected, a message printed saying that the player has been freed (the message should include the name of the player), the `locked_in` attribute of the Player should be set to `False`, and `crystals_collected` should be returned.  If they have insufficient crystals, a message should be printed saying they have insufficient crystals, and `crystals_collected` should be returned.

You can run the code in the code cell below the empty one to test whether your class is working.  Don't forget you must run your code cell after you make any changes for them to take effect.

```python
class Player:
    def __init__(self, name, age, location):
        self.name = name
        self.age = age
        self.location = location
        
        self.locked_in = False
        
    def lock_in(self):
        self.locked_in = True
        
    def be_released(self, crystals_collected):
        if crystals_collected > 0:
            crystals_collected -= 1
            print (f"{self.name} has been freed!")
            self.locked_in = False
            return crystals_collected
        else:
            print ("Insufficient crystals")
            return crystals_collected
```

```python
player_one = Player("Jessica Fletcher", 58, "Cabot Cove")

player_one.lock_in()
if player_one.locked_in == True:
    print (f"{player_one.name} has been locked in!")

current_crystals = 5
print (f"The team has {current_crystals} crystals")
current_crystals = player_one.be_released(current_crystals)
print (f"The team now has {current_crystals} crystals")
```

Build a function named `play_crystal_dome` which takes as inputs the number of crystals that the team has collected, as well as a variable which identifies whether the team is "Low" or "High" skill.  The function should emulate the team collecting silver and gold tickets in the crystal dome.  The number of silver and gold tickets they collect per 5 seconds (each crystal collected gives them 5 seconds of time) will be random, but also partly based on their level of skill.  Specifically, you know that : a) a low-skill team will collect between 10 and 20 silver tickets and between 5 and 15 gold tickets per 5 seconds, b) a high-skill team will collect between 5 and 30 silver tickets and between 20 and 40 gold tickets per 5 seconds.

Once the team's time has run out, messages should be printed informing how many gold and silver tickets were collected by the team.  A final total should be calculated, which deducts the number of silver tickets collected from the number of gold tickets.  If that final total is at least 100, then they win the star prize, otherwise they lose, and an appropriate message should be printed confirming which has happened.

You should test your code by running the function with different numbers of crystals collected (typically teams would win around 5-10 crystals) and with different levels of skill of the team.

```python
def play_crystal_dome(crystals_collected, skill):
    silver_tickets_collected = 0
    gold_tickets_collected = 0
    
    for i in range(crystals_collected):
        if skill == "Low":
            silver_tickets_collected += random.randint(10, 20)
            gold_tickets_collected += random.randint(5, 15)
        else:
            silver_tickets_collected += random.randint(5, 30)
            gold_tickets_collected += random.randint(20, 40)
            
    final_total = gold_tickets_collected - silver_tickets_collected
    
    print (f"You collected {gold_tickets_collected} gold tickets.")
    print (f"You collected {silver_tickets_collected} silver tickets.")
    print (f"This gives you a grand total of {final_total} tickets.")
    
    if final_total >= 100:
        print ("Congratulations - you win the star prize!")
    else:
        print ("Sorry - you lost!")
```

```python
play_crystal_dome(10, "High")
```
